# Perceptron

Exercício simples para treinar um perceptron

## Base
Rede do tipo Perceptron formada por um neurônio com três terminais de entrada, utilizando pesos iniciais w0 = 0.3, w1 = -0.5 e w2 = 0.4, limiar θ = 0.6 e uma taxa de aprendizado = 0.3.

```python
# Dados iniciais
import numpy as np 
# Base de dados
# Classe -1  
A = np.array( [0, 0, 1] ) 
# Classe 1 
B = np.array( [1, 1, 0] ) 
# Pesos iniciais 
Pesos = np.array( [0.3, -0.5, 0.4] ) 
# Taxa de aprendizagem 
eta = 0.6 
# Viés 
bias = 0.3  
```

```python
# Função para a soma ponderada 
# u = x0*w0 +  x1*w1 + x2*w2 -1*θ 
# m = x0*w0 +  x1*w1 + x2*w2
# u = m -1*θ
def somaPonderada(X,W):
  bias = 0.3
  m = np.multiply(X,W)
  u = np.sum(m) -1*bias
  return u 
```

```python
# Função Degrau 
def funcaoAtivacao(u):
  if u >= 0:
    return 1
  else:
    return -1
```

```python
# Função de saída 
def saida(X,W):
  u = somaPonderada(X,W)
  return funcaoAtivacao(u) 
```

```python
# Função para ajustes dos pesos
# Parte cententral do processo de aprendizagem 
# novo wi  →  wi = wi +  η xi (d - y ) 
def ajustaPesos(eta,W,X,bias,d,y):
  e = d - y 
  novoW = W + eta*X*e
  novoB = bias + eta*(-1)*e 
  return novoW, novoB  
```

# Treinamento 
* Realizado manualmente para ilustrar o processo 
* Aprsenta uma entrada e calcula a saída de maneira iterativa 

## Iteração 1

```python
# Entrada 001 saída -1 
d = -1 
y = saida(A,Pesos) 
print(y)
# Ajuste dos pesos 
Pesos,bias = ajustaPesos(eta,Pesos,A,bias,d,y) 
print(Pesos)
print(bias)  

# Entrada 110 saída +1
d = 1 
y = saida(B,Pesos) 
print(y)
# Ajuste dos pesos 
Pesos,bias = ajustaPesos(eta,Pesos,B,bias,d,y) 
print(Pesos)
print(bias) 
```

## Iteração 2

```python
# Entrada 001 saída -1 
d = -1 
y = saida(A,Pesos) 
print(y)
# Ajuste dos pesos 
Pesos,bias = ajustaPesos(eta,Pesos,A,bias,d,y) 
print(Pesos)
print(bias)  

# Entrada 110 saída +1
d = 1 
y = saida(B,Pesos) 
print(y)
# Ajuste dos pesos 
Pesos,bias = ajustaPesos(eta,Pesos,B,bias,d,y) 
print(Pesos)
print(bias) 
```
